```python
# Importing the libraries
import numpy as np # for array operations
import pandas as pd # for working with DataFrames
import matplotlib.pyplot as plt # for data visualization
%matplotlib inline

# scikit-learn modules
from sklearn.model_selection import train_test_split # for splitting the data
from sklearn.metrics import mean_squared_error # for calculating the cost function
from sklearn.ensemble import RandomForestRegressor # for building the model
```

```python
matam_to_hadar1= pd.read_csv("hadar_df.csv")
matam_to_hadar1.head()
```

```python
matam_to_hadar=matam_to_hadar1.head(20000)
```

```python
len(matam_to_hadar)
```

```python
del matam_to_hadar['Unnamed: 0']
del matam_to_hadar['dur']
matam_to_hadar=matam_to_hadar.drop(matam_to_hadar[matam_to_hadar.eq('#VALUE!').any(1)].index) 
matam_to_hadar=matam_to_hadar.drop(matam_to_hadar[matam_to_hadar.eq('#NUM!').any(1)].index) 
matam_to_hadar['start_time']=pd.to_datetime(matam_to_hadar['start_time'], format='%H:%M:%S').dt.hour
#matam_to_hadar['end_time']=pd.to_datetime(matam_to_hadar['end_time'], format='%H:%M:%S').dt.hour
matam_to_hadar.head()
```

```python
one_hot_encoded_data = pd.get_dummies(matam_to_hadar, columns = ['day', 'start_time', 'from','to'])
print(one_hot_encoded_data)
```

```python
one_hot_encoded_data.head()
```

```python
one_hot_encoded_data = one_hot_encoded_data.reset_index(drop=True)
one_hot_encoded_data.dropna(inplace=True)
```

```python
print(one_hot_encoded_data.columns)
```

```python
list(zip( range(0,len(one_hot_encoded_data.columns)),one_hot_encoded_data.columns))
```

```python
Y = one_hot_encoded_data.iloc[:,6].values
X = one_hot_encoded_data.iloc[:,range(7,63)].values
```

```python
# Splitting the dataset into training and testing set (80/20)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 28)
```

```python
# Initializing the Random Forest Regression model with 10 decision trees
model = RandomForestRegressor(n_estimators = 60, random_state = 1234)

# Fitting the Random Forest Regression model to the data
model.fit(x_train, y_train)
```

```python
# Predicting the target values of the test set
y_pred = model.predict(x_test)

# RMSE (Root Mean Square Error)
rmse = float(format(np.sqrt(mean_squared_error(y_test, y_pred)), '.3f'))
print("\nRMSE: ", rmse)
```
